```python
import yfinance as yf
import pandas as pd

# Create a Ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Extract Tesla stock history data
tesla_data = tesla.history(period="max")

# Reset the index and display the first five rows
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

# Get Tesla revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
response = requests.get(url, headers=headers)

# Check if request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")

    # Extract all tables
    tables = soup.find_all("table")

    # Debug: Print the first few characters of each table to find the correct one
    for i, table in enumerate(tables):
        print(f"Table {i}: {table.text[:200]}")  # Print first 200 characters

    # Extract tables using pandas
    dfs = pd.read_html(response.text)  # Extract tables directly

    # Debug: Check all extracted tables
    for i, df in enumerate(dfs):
        print(f"DataFrame {i}:")
        print(df.head())

    # Select the correct dataframe (adjust index if needed)
    tesla_revenue_df = dfs[0]  # Change index if needed

    # Clean the dataframe
    tesla_revenue_df.columns = ["Date", "Revenue"]
    tesla_revenue_df = tesla_revenue_df.dropna()
    tesla_revenue_df["Revenue"] = tesla_revenue_df["Revenue"].astype(str).str.replace(",", "").str.replace("$", "").astype(float)

    # Display the first five rows
    print(tesla_revenue_df.head())

else:
    print(f"Request failed with status code {response.status_code}")
```

```python
# Create a Ticker object for GameStop
gamestop = yf.Ticker("GME")

# Extract GameStop stock history data
gamestop_data = gamestop.history(period="max")

# Reset the index and display the first five rows
gamestop_data.reset_index(inplace=True)
gamestop_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define headers to mimic a real browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

# Get GameStop revenue data
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
response = requests.get(url, headers=headers)

# Check if request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")

    # Extract tables
    tables = soup.find_all("table")

    # Debug: Print a preview of each table
    for i, table in enumerate(tables):
        print(f"Table {i}: {table.text[:200]}")  # Print first 200 characters

    # Extract tables using Pandas
    dfs = pd.read_html(response.text)  # Extract tables directly

    # Debug: Check all extracted tables
    for i, df in enumerate(dfs):
        print(f"DataFrame {i}:")
        print(df.head())

    # Select the correct dataframe (adjust index if needed)
    gamestop_revenue_df = dfs[0]  # Change index if needed

    # Clean the dataframe
    gamestop_revenue_df.columns = ["Date", "Revenue"]
    gamestop_revenue_df = gamestop_revenue_df.dropna()
    gamestop_revenue_df["Revenue"] = gamestop_revenue_df["Revenue"].astype(str).str.replace(",", "").str.replace("$", "").astype(float)

    # Display the first five rows
    print(gamestop_revenue_df.head())

else:
    print(f"Request failed with status code {response.status_code}")
```

```python
import matplotlib.pyplot as plt

# Create Tesla stock and revenue dashboard
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot stock price
ax1.set_xlabel("Year")
ax1.set_ylabel("Stock Price (USD)", color="tab:blue")
ax1.plot(tesla_data["Date"], tesla_data["Close"], color="tab:blue", label="Tesla Stock Price")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Create a second y-axis
ax2 = ax1.twinx()
ax2.set_ylabel("Revenue (Million USD)", color="tab:red")
ax2.plot(tesla_revenue_df["Date"], tesla_revenue_df["Revenue"], color="tab:red", label="Tesla Revenue")
ax2.tick_params(axis="y", labelcolor="tab:red")

fig.tight_layout()
plt.title("Tesla Stock Price vs Revenue Over Time")
plt.show()
```
